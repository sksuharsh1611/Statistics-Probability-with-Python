```python
# Ans 1:
'''
The three measures of central tendency are: 
1. Mean
2. Median 
3. Mode
'''
```

```python
# Ans 2:
'''
1. Mean: 
> The mean is the average of a set of data.
> It is calculated by adding up all the values in the data set and dividing by the total number of values. 
> The mean is sensitive to extreme values or outliers.

2. Median:
> The median is the middle value in a set of data when the values are arranged in order. 
> If there is an even number of values, the median is the average of the two middle values. 
> The median is not affected by extreme values or outliers.

3. Mode: 
> The mode is the value that occurs most frequently in a set of data. 
> A set of data can have one mode; 
> multiple modes (when two or more values appear with the same frequency); 
> or no mode (when no value is repeated in the data set).

'''
'''
Note: 
* These measures are important because they provide a representative value that summarizes the data in a meaningful way. 
* The choice of measure depends on the type of data and the research question being addressed. 
* It is important to select the appropriate measure of central tendency to avoid distorting the results or drawing incorrect conclusions.
'''
```

```python
# Ans 3:
import numpy as np
from scipy import stats

list1 = np.array([178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5])

print("Mean:", np.mean(list1))
print("Median:", np.median(list1))
print("Mode:", stats.mode(list1,keepdims=False))
```

```python
# Ans 8:
'''
>> Skewness refers to the degree of asymmetry in the distribution of data.

>> Skewness can be positive, negative, or zero.

>> Positive skewness means the tail is longer on the right side, 
   Negative skewness means the tail is longer on the left side, 
   Zero skewness means the data are symmetric.
   
>> Skewed data can affect the interpretation of statistical measures such as the mean and standard deviation.

>> Skewed data may require different statistical techniques or transformations to be properly analyzed.
'''
```

```python
# Ans 9:
'''
Since, in a right-skewed distribution, the tail of the distribution is longer on the right side.

>> The mean is affected by extreme values, and will be pulled towards the right by the larger values.
>> The median is resistant to the influence of extreme values and will be closer to the lower end of the dataset.
>> Therefore, the median will typically be less than the mean in a right-skewed distribution.
'''
```

```python
# Ans 10:
'''
Covariance and correlation are both measures of the relationship between two variables, but they differ in several ways:

1. Scale dependence: 
            Covariance is a scale-dependent measure, 
      while correlation is a scale-independent measure. 

            Covariance can be affected by the scale of the variables,
      while correlation is not affected by the scale.

2. Range: 
        Covariance can take any value, 
  while correlation always ranges between -1 and 1.

3. Interpretation: 
        Covariance measures the direction and strength of the linear relationship between two variables. 
        However, it doesn't provide information about the strength of the relationship. 
        
        Correlation, on the other hand, provides information about both the direction and the strength of the linear relationship between two variables.

4. Standardization: 
        Correlation is a standardized measure of covariance.
    It is obtained by dividing the covariance by the product of the standard deviations of the two variables.
'''

'''
These measures used in statistical analysis as following points:

-> A positive covariance indicates that the variables tend to move in the same direction, while a negative covariance indicates they tend to move in opposite directions.
-> However, covariance values can be difficult to interpret, as they depend on the scale of the variables.

-> Correlation measures the strength and direction of the linear relationship between two variables.
-> Correlation values range from -1 to 1, with -1 indicating a perfect negative relationship, 0 indicating no relationship, and 1 indicating a perfect positive relationship.
-> Correlation is more commonly used than covariance because it is easier to interpret and compare across different datasets and variables.
-> Correlation is often used to identify trends and patterns in data, as well as to test hypotheses and make predictions.

'''
```

```python
# Ans 11:
'''
The formula for calculating the sample mean is:

sample mean = sum of all observations / number of observations

To calculate the sample mean, we add up all the values in the dataset and divide by the number of observations. 
'''
# Here's an example calculation for a dataset:
# Define a list of observations
age = [5, 10, 15, 20, 25]

# Calculate the sample mean
sample_mean = sum(age) / len(age)

# Print the result
print("Sample mean: ", sample_mean)
```

```python
# Ans 12:
'''
For a normal distribution data, the three measures of central tendency: mean, median, and mode are all equal to each other. 
This means that the mean, median, and mode are located at the same point on the horizontal axis of the normal distribution curve.

>> This is because a normal distribution is symmetric around its mean, which is also the midpoint of the distribution. 
>> The median is also located at this midpoint, because it is the point that divides the distribution into two equal halves. 
>> Additionally, the mode is also equal to the mean and median, as it represents the most frequently occurring value in the distribution, which is also at the midpoint of the curve in a normal distribution.
'''
# for example:
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Generate a normally distributed dataset with mean 10 and standard deviation 2
data = np.random.normal(10, 2, 1000)

# Calculate the mean, median, and mode
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data,keepdims=True)[0][0]

# Create a histogram of the data
plt.hist(data, bins=30)

# Add vertical lines for the mean, median, and mode
plt.axvline(mean, color='r', label='Mean')
plt.axvline(median, color='g', label='Median')
plt.axvline(mode, color='b', label='Mode')

# Add a legend and axis labels
plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Ans 13:
'''
The key differences between covariance and correlation:

>> Covariance is a measure of how two variables change together,
   while correlation is a measure of how strong and consistent the relationship is between two variables.

>> Covariance has units of measurement,
   while correlation is a dimensionless measure that ranges from -1 to 1.

>> Covariance can take on any value, 
   while correlation always ranges from -1 to 1.

>> Correlation is standardized, meaning it is not affected by differences in the units of measurement of the two variables,
   while covariance is not standardized and can be affected by differences in scale.

>> Correlation is calculated by dividing the covariance by the product of the standard deviations of the two variables.
'''
```

```python
# Ans 14:
'''
Outliers are values that are significantly different from the other values in a dataset. 
They can have a substantial effect on the measures of central tendency and dispersion. 

# Here are some examples of how outliers can affect these measures:
===============================================================================================================================
1. Mean: Outliers can significantly affect the mean, which is the average value of a dataset. 

For example:
 If we have a dataset with values {1, 2, 3, 4, 100}, the mean is (1+2+3+4+100)/5 = 22. 
 If we remove the outlier value of 100, the mean becomes (1+2+3+4)/4 = 2.5, which is much lower than the previous mean.

===============================================================================================================================

2. Median: The median is the value that separates the higher half from the lower half of a dataset. 
           Outliers may have less effect on the median than on the mean, especially if there are few of them. 
        
For example:
 If we have a dataset with values {1, 2, 3, 4, 100}, the median is 3, which is not affected by the outlier value of 100.

===============================================================================================================================

3. Mode: The mode is the value that appears most frequently in a dataset. 
         Outliers may not affect the mode at all if they are far from the other values. 
        
For example:
 If we have a dataset with values {1, 2, 3, 4, 100}, the mode is 1, which is not affected by the outlier value of 100.

================================================================================================================================

4. Range: The range is the difference between the maximum and minimum values in a dataset. 
          Outliers can significantly affect the range, especially if they are much higher or lower than the other values. 
          
For example:
  If we have a dataset with values {1, 2, 3, 4, 100}, the range is 100-1 = 99. 
  If we remove the outlier value of 100, the range becomes 4-1 = 3, which is much lower than the previous range.

=================================================================================================================================
5. Standard deviation: The standard deviation is a measure of how spread out the values in a dataset are. 
                       Outliers can significantly affect the standard deviation, especially if they are far from the other values. 
                       
For example:
  If we have a dataset with values {1, 2, 3, 4, 100}, the standard deviation is 41.05. 
  If we remove the outlier value of 100, the standard deviation becomes 1.29, which is much lower than the previous standard deviation.

===================================================================================================================================
'''
'''
:*: Important Note :*: 

In summary, outliers can have a significant effect on measures of central tendency and dispersion, especially on the mean and standard deviation. 
It is important to identify and handle outliers appropriately to avoid skewing the results of statistical analyses.

'''
```
