```python
# Ans 1:
Differences between a t-test and a z-test:

1. Population variance known or unknown: The z-test is used when the population variance is known, 
                                        while the t-test is used when the population variance is unknown.

2. Sample size: The z-test requires a large sample size (typically n > 30), 
                while the t-test can be used with smaller sample sizes.

3. Type of data: The z-test is used to test hypotheses about population means when the data is normally distributed, 
                 while the t-test can be used with non-normal data if the sample size is large enough.

4. Test statistic: The z-test uses the z-statistic, which is the number of standard deviations a sample mean is from the population mean, 
                   while the t-test uses the t-statistic, which is the ratio of the difference between the sample mean and the population mean to the standard error.

5. Confidence intervals: The confidence interval for a z-test is calculated using the z-distribution, 
                         while the confidence interval for a t-test is calculated using the t-distribution.

Overall, the t-test is a more flexible test that can be used with smaller sample sizes and non-normal data, 
   while the z-test is a more specialized test that requires a larger sample size and normal data.
    
'''
Example Scenario for a t-test:
Suppose you want to know whether there is a significant difference between the average weight of apples produced by two different orchards. 
You take a random sample of 25 apples from each orchard and measure their weights. 
The population variance of the apple weights is unknown. 
In this case, you would use a two-sample t-test to compare the means of the two samples and determine whether there is a statistically significant difference between the two orchards.

Example Scenario for a z-test:
Suppose you want to know whether the average height of all American adult men is greater than 6 feet.
You take a random sample of 1000 adult men and measure their heights. 
The population variance of the height of American adult men is known to be 4 inches. 
In this case, you would use a one-sample z-test to determine whether the sample mean is significantly different from 6 feet. 
Since the population variance is known and the sample size is large enough, a z-test can be used in this case.
'''
```

```python
# Ans 2:
'''
Differences between one-tailed and two-tailed tests:

1. Direction of the hypothesis: A one-tailed test is used when the hypothesis specifies the direction of the difference between two groups or variables, 
                                while a two-tailed test is used when the hypothesis does not specify the direction of the difference.

2. Critical region: The critical region for a one-tailed test is located entirely in one tail of the distribution, 
                    while the critical region for a two-tailed test is split between the two tails of the distribution.

3. Alpha level: The alpha level (or level of significance) is typically divided between the two tails for a two-tailed test,
                while the alpha level is concentrated in one tail for a one-tailed test.

4. Power: A one-tailed test is more powerful than a two-tailed test, since it concentrates all of the alpha level in one tail of the distribution. 
         However, a one-tailed test can also be more biased and less generalizable than a two-tailed test, since it assumes a specific direction of the effect.

5. Interpretation of results: The interpretation of the results of a one-tailed test is straightforward, since it only tests for a specific direction of the effect. 
                              The interpretation of the results of a two-tailed test is more nuanced, since it tests for the possibility of an effect in either direction.
'''
```

```python
# Ans 3:
'''
Type 1 error:
A Type 1 error occurs when we reject the null hypothesis when it is actually true. 
In other words, we conclude that there is a significant difference between two groups or variables when there is actually no difference. 
The probability of a Type 1 error is denoted by alpha (α) and is typically set at 0.05 or 0.01.

Type 2 error: 
A Type 2 error occurs when we fail to reject the null hypothesis when it is actually false. 
In other words, we conclude that there is no significant difference between two groups or variables when there is actually a difference. 
The probability of a Type 2 error is denoted by beta (β) and is influenced by several factors, such as sample size, effect size, and alpha level.


Examples of Type 1 and Type 2 errors are:

Type 1 error: Rejecting the null hypothesis that a new drug has no side effects, when in fact it does not have side effects. 
              This can lead to unnecessary withdrawal of a potentially beneficial drug from the market.

Type 2 error: Failing to reject the null hypothesis that a new drug has no side effects, when in fact it does have side effects. 
              This can lead to the release of a drug with harmful side effects onto the market.

Overall, Type 1 and Type 2 errors represent the two types of errors that can occur in hypothesis testing. 
These errors are a natural part of the hypothesis testing process, and researchers must carefully balance the risks of making each type of error when designing and interpreting their studies.
'''
```

```python
# Ans 4:
Bayes's theorem: Bayes's theorem is a mathematical formula used to update the probability of a hypothesis based on new evidence.

>> Prior probability: The prior probability is the initial probability of the hypothesis before new evidence is taken into account.

>> Likelihood: The likelihood is the probability of the new evidence given the hypothesis.

>> Posterior probability: The posterior probability is the updated probability of the hypothesis after the new evidence is taken into account.

Formula: Bayes's theorem can be expressed as P(A|B) = P(B|A) * P(A) / P(B), 
        where P(A|B) is the posterior probability of A given B, 
              P(B|A) is the likelihood of B given A, 
              P(A) is the prior probability of A, 
              P(B) is the marginal probability of B.

Example: Suppose we want to diagnose a patient with a rare disease. 
         The prior probability of the patient having the disease is 1%, based on the prevalence of the disease in the general population. 
         We conduct a test that has a sensitivity of 95% and a specificity of 90%. 
         If the test is positive, what is the updated probability that the patient has the disease?

--> Solution: Using Bayes's theorem, we can calculate the posterior probability as follows:

P(Disease|Positive test) = P(Positive test|Disease) * P(Disease) / P(Positive test) 
                         = 0.95 * 0.01 / (0.95 * 0.01 + 0.10 * 0.99) 
                         = 8.7%. 
Thus, the updated probability that the patient has the disease is 8.7%, which is higher than the prior probability of 1% but still relatively low due to the low prevalence of the disease.

Overall, Bayes's theorem is a powerful tool for updating the probability of a hypothesis based on new evidence, and can be used in a wide range of applications, from medical diagnosis to machine learning.
```

```python
# Ans 5:
'''
A confidence interval is a statistical range of values that is likely to contain the true value of an unknown population parameter with a certain level of confidence.

Formula: A confidence interval can be calculated using the formula:
        CI = x ± z*SE, 
        where CI is the confidence interval,
        x is the sample statistic,
        z is the critical value from the standard normal distribution based on the level of confidence, 
        SE is the standard error of the sample statistic.
'''
For example, 
A pollster wants to estimate the proportion of voters who will support a particular candidate in an upcoming election.
They take a random sample of 500 voters and find that 55% of them support the candidate. 
They construct a 95% confidence interval for the true proportion of voters who support the candidate using the formula: 
    CI = 0.55 ± 1.96 * sqrt(0.55 * 0.45 / 500). 
    --> The resulting confidence interval is 0.49 to 0.61. 
    --> This means that we are 95% confident that the true proportion of voters who support the candidate lies between 49% and 61%. 
The margin of error for this confidence interval is 6%
```

```python
# Ans 6:
Problem:
Suppose a rare disease affects 1 in 100,000 people in a certain population.
A test has been developed to diagnose the disease, but it is not perfect. 
The test correctly identifies 99% of people who have the disease (true positive rate) but also wrongly classifies 1% of healthy people as having the disease (false positive rate). 
If a randomly selected person tests positive for the disease, what is the probability that they actually have the disease?

Solution:
Let A be the event that a person has the disease, and B be the event that a person tests positive for the disease. 
We want to find P(A|B), the probability of having the disease given a positive test result.

Using Bayes' theorem, we have: P(A|B) = P(B|A) * P(A) / P(B)

From the problem statement, we know that:

>> P(A) = 1/100000

>> P(B|A) = 0.99 (true positive rate)

>> P(B|not A) = 0.01 (false positive rate)

#To calculate P(B), we use the law of total probability:

P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)

where P(not A) = 1 - P(A) is the complement of the event A.

Therefore, we have:

P(B) = 0.99 * 1/100000 + 0.01 * 99999/100000 = 0.010098

Finally, we can plug in these values into Bayes' theorem to get:

P(A|B) = 0.99 * 1/100000 / 0.010098 = 0.0098

Thus, the probability that a person who tests positive for the disease actually has the disease is about 0.98%, or less than 1%. 
This result highlights the importance of considering both the sensitivity and specificity of a test when interpreting its results.
```

```python
# Ans 8:
The margin of error in a confidence interval is the amount by which a point estimate (e.g., sample mean or proportion) can vary from the true population parameter, given a certain level of confidence. 

Sample size affects the margin of error in that larger samples tend to have smaller margins of error, 
    while smaller samples tend to have larger margins of error. 
This is because larger samples provide more information and reduce the impact of random variation in the data.

For example, 
Consider a poll that wants to estimate the proportion of people in a city who support a new policy proposal, with a 95% confidence level and a margin of error of 3%. 
If the poll is conducted with a sample size of 1000, the margin of error would be approximately ±3%, which means that we can be 95% confident that the true proportion of supporters in the population is within the range of 47% to 53%. 
However, if the same poll is conducted with a sample size of 2000, the margin of error would be approximately ±2%, which means that we can be 95% confident that the true proportion of supporters in the population is within the range of 48% to 52%. 
In this scenario, the larger sample size has reduced the margin of error from 3% to 2%.
```

```python
# Ans 10:
import numpy as np
import scipy.stats as stats

# Define the null hypothesis: the drug is not significantly effective
null_hypothesis = "The drug is not significantly effective"

# Define the alternative hypothesis: the drug is significantly effective
alt_hypothesis = "The drug is significantly effective"

# Set the level of significance and degrees of freedom
alpha = 0.05
df = 49  # degrees of freedom = sample size - 1

# Define the sample data
sample_mean = 6
pop_mean = 0  # since the null hypothesis assumes no effect
pop_std = 2.5
sample_data = np.random.normal(sample_mean, pop_std, size=50)

# Calculate the t-statistic and p-value
t_stat, p_val = stats.ttest_1samp(sample_data, pop_mean, nan_policy='omit')

# Interpret the results
if p_val < alpha:
    print(f"Reject the null hypothesis. {alt_hypothesis}. t-statistic: {t_stat:.2f}, p-value: {p_val:.4f}")
else:
    print(f"Fail to reject the null hypothesis. {null_hypothesis}. t-statistic: {t_stat:.2f}, p-value: {p_val:.4f}")
```

```python
# Ans 12:
import numpy as np
from scipy.stats import t

# Sample A data
sample_a_mean = 85
sample_a_std = 6
sample_a_n = 30

# Sample B data
sample_b_mean = 82
sample_b_std = 5
sample_b_n = 30

# Calculate the pooled standard deviation
pooled_std = np.sqrt(((sample_a_n - 1)*(sample_a_std**2) + (sample_b_n - 1)*(sample_b_std**2)) / (sample_a_n + sample_b_n - 2))

# Calculate the t-statistic
t_statistic = (sample_a_mean - sample_b_mean) / (pooled_std * np.sqrt(1/sample_a_n + 1/sample_b_n))

# Calculate the critical value for the two-tailed t-test with alpha = 0.01
t_critical = t.ppf(1-0.01/2, sample_a_n + sample_b_n - 2)

# Compare t_statistic with the critical value
if np.abs(t_statistic) > t_critical:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
# Ans 14:
import numpy as np
from scipy.stats import t

# Sample statistics
n = 30
x_bar = 0.25
s = 0.05

# Null hypothesis: caffeine has no significant effect on reaction time
mu0 = 0

# Calculate the t-statistic
t_stat = (x_bar - mu0) / (s / np.sqrt(n))

# Calculate the p-value
p_value = 2 * (1 - t.cdf(np.abs(t_stat), n - 1))

# Set the significance level
alpha = 0.1

# Compare the p-value to the significance level
if p_value < alpha:
    print("Reject the null hypothesis. Caffeine has a significant effect on reaction time.")
else:
    print("Fail to reject the null hypothesis. Caffeine has no significant effect on reaction time.")
```
